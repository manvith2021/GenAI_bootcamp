```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

reviews = []
base_url = "https://www.amazon.in/product-reviews/B08N5WRWNW?pageNumber="

for page in range(1, 6):  # scrape first 5 pages
    response = requests.get(base_url + str(page), headers={"User-Agent":"Mozilla/5.0"})
    soup = BeautifulSoup(response.content, "html.parser")
    
    for block in soup.find_all("span", {"class": "review-body"}):
        text = block.get_text(strip=True)
        reviews.append(text)
    
    time.sleep(2)  # polite delay

df = pd.DataFrame(reviews, columns=["review_text"])
df.to_csv("amazon_reviews.csv", index=False)
print("Scraped", len(reviews), "reviews")
```
